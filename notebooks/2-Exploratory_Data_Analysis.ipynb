### 4.1 Year Month<a id='4.1_Year_Month'></a>

Note that the column `Year Month` is currently of type `int64`. `matplotlib` has trouble plotting period data, and so instead, we will use a bit of a hack: every distinct `Year Month` will have an associated and ordered numeric representation in column `Period`. That is, January 2018 will be represented by the number 1, February 2018 by the number 2, and so on. Then, when plotting, we will utilize the numeric column data but display the `Year Month` data in the axis labels.

We'll also convert `Year Month` to type `str`.

```python
df['Year Month'].dtype
```

```python
# convert it to type string, then datetime, then to type Period
#data['Year Month'] = pd.to_datetime( data['Year Month'].astype(str), format='%Y%m').dt.to_period("M")
df['Year Month'] = df['Year Month'].astype(str)
```

```python
# create mapping from 'Year Month' to 'Period'
periodKeys = np.sort(df['Year Month'].unique()).tolist()
periodMapper = dict(zip(periodKeys,np.arange(1,len(periodKeys)+1,1)))
periodMapper
```

```python
df['Period'] = df['Year Month'].apply(lambda x: periodMapper[x])
```

```python
# define useful function to create scatterplots of Median Sale Price against desired columns
# show regression and accept R coefficient values
def scatterplotswithreg(columns, ncol=None, figsize=(15, 8), annotateR=False, rValues={}):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.2, hspace=0.4)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        #ax.scatter(x = col, y = 'Median Sale Price', data=df, alpha=0.5)
        g = sns.regplot(x = col, y = 'Median Sale Price', data=df, ax=ax)
        if annotateR:
            g.annotate("r = {:.2f}".format(rValues[col]), xy=(.7,.9), xycoords=ax.transAxes)
        ax.set(xlabel=col, ylabel='Housing price')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

## 6 Statistical Testing of Time Series aspect<a id='6_Statistical_Testing'></a>

We want to confirm that we can disregard the Time Series aspect of the data. To do this, we will do the following:
  1. Perform a one-sample t-test on the year-over-year price difference by Neighborhood.
  2. Perform a one-sample t-test on the month-over-month price difference by Neighborhood.
  
In each case, we will use a null hypothesis that the mean of such a difference is zero.

```python
# create scaled down dataframe to setup price differences across periods
df_period_neighborhood = df.loc[:,['Year Month','Neighborhood','Median Sale Price','Period']]
df_period_neighborhood['Next Monthly Period'] = df_period_neighborhood['Period'] + 1
df_period_neighborhood['Next Yearly Period'] = df_period_neighborhood['Period'] + 12
df_period_neighborhood.head()
```

### 6.1 Year-over-year comparison<a id='6.1_Year-over-year_comparison'></a>

```python
# compare year over year
df_diff_yearly = pd.merge(df_period_neighborhood, df_period_neighborhood, how='inner', 
         left_on=['Neighborhood','Next Yearly Period'], right_on=['Neighborhood','Period'],
         suffixes=('_prev','_curr')).drop(columns=['Next Monthly Period_prev','Next Monthly Period_curr','Next Yearly Period_prev','Next Yearly Period_curr'])
df_diff_yearly['Year-over-year Price Change'] = df_diff_yearly['Median Sale Price_curr'] - df_diff_yearly['Median Sale Price_prev']
df_diff_yearly.drop(columns=['Median Sale Price_prev','Median Sale Price_curr'], inplace=True)
```

```python
# plot year-over-year with regression
#_ = sns.lmplot(x='Period_curr', y='Year-over-year Price Change', data=df_diff_yearly, height=14, hue='Neighborhood')
g = sns.jointplot(x='Period_curr', y='Year-over-year Price Change', data=df_diff_yearly, height=12, kind='hex')
g.annotate(stats.pearsonr)
_ = sns.regplot(x='Period_curr', y='Year-over-year Price Change', data=df_diff_yearly, ax=g.ax_joint, scatter=False)
```

Let's declare our hypotheses:
  * Null hypothesis H0: Year-over-year median housing sales price differences have a mean that is __equal__ to zero.
  * Alternative hypothesis H1: Year-over-year median housing sales price differences have a mean that is __not equal__ to zero.
  
We will use a one-sample t-test.

```python
tStat, pValue = stats.ttest_1samp(df_diff_yearly['Year-over-year Price Change'], popmean=0)
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) 
```

Since the P-Value is not low, 0.446 in this case, we fail to reject the Null Hypothesis.

### 6.2 Month-over-month comparison<a id='6.2_Month-over-month_comparison'></a>

```python
# compare month over month
df_diff_monthly = pd.merge(df_period_neighborhood, df_period_neighborhood, how='inner', 
         left_on=['Neighborhood','Next Monthly Period'], right_on=['Neighborhood','Period'],
         suffixes=('_prev','_curr')).drop(columns=['Next Monthly Period_prev','Next Monthly Period_curr','Next Yearly Period_prev','Next Yearly Period_curr'])
df_diff_monthly['Month-over-month Price Change'] = df_diff_monthly['Median Sale Price_curr'] - df_diff_monthly['Median Sale Price_prev']
df_diff_monthly.drop(columns=['Median Sale Price_prev','Median Sale Price_curr'], inplace=True)
```

```python
# plot month-over-month with regression
#_ = sns.lmplot(x='Period_curr', y='Month-over-month Price Change', data=df_diff_monthly, height=14, hue='Neighborhood')
g = sns.jointplot(x='Period_curr', y='Month-over-month Price Change', data=df_diff_monthly, height=12, kind='hex')
g.annotate(stats.pearsonr)
_ = sns.regplot(x='Period_curr', y='Month-over-month Price Change', data=df_diff_monthly, ax=g.ax_joint, scatter=False)
```

Let's declare our hypotheses:
  * Null hypothesis H0: Month-over-month median housing sales price differences have a mean that is __equal__ to zero.
  * Alternative hypothesis H1: Month-over-month median housing sales price differences have a mean that is __not equal__ to zero.
  
We will use a one-sample t-test.

```python
tStat, pValue = stats.ttest_1samp(df_diff_monthly['Month-over-month Price Change'], popmean=0)
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat)) 
```

Since the P-Value is not low, 0.97 in this case, we fail to reject the Null Hypothesis.

### 6.3 Conclusion<a id='6.3_Conclusion'></a>

In conclusion, with such high P-Values, we fail to reject the Null Hypothesis, and we will disregard the Time Series aspect of the data.

```python
# let's drop the Period column since we've decided to disregard the Time Series aspect of the data
df.drop(columns=['Period'], inplace=True)
```
